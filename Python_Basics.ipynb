##Python Basics Questions

###1. What is Python, and why it is popular?
Python is a programming language that's used to build websites, software, and automate tasks. It's a general-purpose language that's easy to learn and use. It is popular because of its:
-Readability
-Flexibility
-Compatibility
-Open source

###2. What is an interpreter in Python?
In Python, an interpreter is a program that reads and executes Python code line by line, translating it into machine-readable instructions and then executing them.
The Python interpreter acts as a bridge between the Python code you write and the computer's hardware. It takes your code, which is written in a human-readable format, and converts it into a form that the computer can understand and execute.


###3. What are pre-defined keywords in Python?
In Python, "predefined" refers to built-in functions, data types, and constants that are automatically available for use without requiring explicit import statements.
For example: print, strint(), len(), type(), input(), int(), float(), str(), list(), dict(), range(), open() and many more.

###4. Can keywords be used as variable names?
No, you cannot use Python keywords as variable names because they are reserved words with specific meanings in the language syntax.

###5. What is mutability in Python?
In Python, mutability refers to an object's ability to be changed after it's created; mutable objects can be modified in place, while immutable objects cannot.

###6. Why are lists mutable, but tuples are immutable?
Tuples are immutable because the items in tuples cannot be modified but the items in lists can be modified, that's why it is called mutable.

###7.What is the difference between “==” and “is” operators in Python?
In Python, == checks for value equality (if two objects have the same content), while is checks for object identity (if two variables point to the same object in memory).

###8. What are logical operators in Python?
In Python, logical operators, and, or, and not, are used to combine or reverse boolean expressions, returning True or False based on the truth values of the operands.

###9. What is type casting in Python?
In Python, type casting (also known as type conversion) is the process of converting a variable's data type to another type, allowing you to perform operations or manipulate data more flexibly.

###10. What is the difference between implicit and explicit type casting?
Implicit type casting, also known as widening, occurs automatically by the compiler when converting a smaller data type to a larger one, while explicit type casting, or narrowing, requires the programmer to use a cast operator to convert a larger type to a smaller one.

###11. What is the purpose of conditional statements in Python?
Conditional statements in Python, like if, elif, and else, allow a program to make decisions and execute different code blocks based on whether a condition is true or false, enabling dynamic and flexible program behavior.

###12. How does the elif statement work?
The “elif” keyword in Python, stands for “else if”. It can be used in conditional statements to check for multiple conditions. For example, if the first condition is false, it moves on to the next “elif” statement to check if that condition is true.

###13. What is the difference between for and while loops?
The main difference between a for loop and a while loop is that a for loop is used when you know the number of iterations beforehand, while a while loop is used when the number of iterations is unknown and depends on a condition.

###14. Describe a scenario where a while loop is more suitable than a for loop.
In contrast to a 'for' loop, which is generally used when the number of iterations is known beforehand, a 'while' loop is ideal for situations where the number of iterations is not predetermined and depends on some dynamic condition.

```python
#Q.7. Write a program that performs basic arithmetic operations on two user-input numbers.


def perform_arithmetic(num1, num2, operation):
    """
    Performs basic arithmetic operations on two numbers.

    Args:
        num1: The first number.
        num2: The second number.
        operation: The arithmetic operation to perform (+, -, *, /).

    Returns:
        The result of the operation, or an error message if the operation is invalid or division by zero is attempted.
    """

    if operation == "+":
        return num1 + num2
    elif operation == "-":
        return num1 - num2
    elif operation == "*":
        return num1 * num2
    elif operation == "/":
        if num2 == 0:
            return "Error: Division by zero"
        else:
            return num1 / num2
    else:
        return "Error: Invalid operation"

def get_user_input():
    """
    Gets user input for two numbers and an arithmetic operation.

    Returns:
        A tuple containing the two numbers and the operation.
    """

    while True:
        try:
            num1 = float(input("Enter the first number: "))
            break
        except ValueError:
            print("Invalid input. Please enter a number.")

    while True:
        try:
            num2 = float(input("Enter the second number: "))
            break
        except ValueError:
            print("Invalid input. Please enter a number.")

    while True:
        operation = input("Enter the operation (+, -, *, /): ")
        if operation in ("+", "-", "*", "/"):
            break
        else:
            print("Invalid operation. Please enter +, -, *, or /.")

    return num1, num2, operation

# Main program
num1, num2, operation = get_user_input()
result = perform_arithmetic(num1, num2, operation)
print("Result:", result)
```

```python
#Q.8. Write a program to demonstrate the use of logical operators.


def demonstrate_logical_operators(a, b, c):
    """
    Demonstrates the use of logical operators (and, or, not) in Python.

    Args:
        a: A boolean value.
        b: A boolean value.
        c: A boolean value.
    """

    print(f"a: {a}, b: {b}, c: {c}")

    # AND operator
    result_and = a and b
    print(f"a and b: {result_and}")

    # OR operator
    result_or = a or b
    print(f"a or b: {result_or}")

    # NOT operator
    result_not_a = not a
    print(f"not a: {result_not_a}")

    # Combined logical operations
    result_combined = (a and b) or (not c)
    print(f"(a and b) or (not c): {result_combined}")

    result_combined2 = a and (b or not c)
    print(f"a and (b or not c): {result_combined2}")

# Example usage:
demonstrate_logical_operators(True, False, True)
print("-" * 20)
demonstrate_logical_operators(False, True, False)
print("-" * 20)
demonstrate_logical_operators(True, True, False)
print("-" * 20)
demonstrate_logical_operators(False, False, True)
```

```python
#Q.9. Write a Python program to convert user input from string to integer, float, and boolean types


def convert_user_input():
    """
    Converts user input from string to integer, float, and boolean types.
    """

    user_input = input("Enter a value: ")

    try:
        # Convert to integer
        integer_value = int(user_input)
        print(f"Integer: {integer_value}")
    except ValueError:
        print("Cannot convert to integer.")

    try:
        # Convert to float
        float_value = float(user_input)
        print(f"Float: {float_value}")
    except ValueError:
        print("Cannot convert to float.")

    # Convert to boolean (case-insensitive)
    if user_input.lower() == "true":
        boolean_value = True
        print(f"Boolean: {boolean_value}")
    elif user_input.lower() == "false":
        boolean_value = False
        print(f"Boolean: {boolean_value}")
    else:
        print("Cannot convert to boolean.")

# Example usage:
convert_user_input()

print("\nTesting some edge cases:")
#Test cases with spaces and numbers
test_inputs = ["  123  ", "4.56  ", "  True  ", "  False  ", "hello", "12.3.4", "true1"]

for test in test_inputs:
  print(f"\nTesting input: '{test}'")
  try:
      integer_value = int(test.strip()) #remove surrounding spaces
      print(f"Integer: {integer_value}")
  except ValueError:
      print("Cannot convert to integer.")

  try:
      float_value = float(test.strip())
      print(f"Float: {float_value}")
  except ValueError:
      print("Cannot convert to float.")

  if test.strip().lower() == "true":
      boolean_value = True
      print(f"Boolean: {boolean_value}")
  elif test.strip().lower() == "false":
      boolean_value = False
      print(f"Boolean: {boolean_value}")
  else:
      print("Cannot convert to boolean.")
```

```python
#Q.10.  Write code to demonstrate type casting with list elements.


def type_casting_list_elements(mixed_list):
    """
    Demonstrates type casting with list elements.

    Args:
        mixed_list: A list containing elements of different types.
    """

    print("Original list:", mixed_list)

    # Cast elements to integers (if possible)
    integer_list = []
    for element in mixed_list:
        try:
            integer_list.append(int(element))
        except (ValueError, TypeError):
            integer_list.append(None)  # Or handle the error differently

    print("List cast to integers:", integer_list)

    # Cast elements to floats (if possible)
    float_list = []
    for element in mixed_list:
        try:
            float_list.append(float(element))
        except (ValueError, TypeError):
            float_list.append(None)  # Or handle the error differently

    print("List cast to floats:", float_list)

    # Cast elements to strings
    string_list = []
    for element in mixed_list:
        string_list.append(str(element))

    print("List cast to strings:", string_list)

    #Cast to boolean
    boolean_list = []
    for element in mixed_list:
      boolean_list.append(bool(element))

    print("List cast to booleans:", boolean_list)

# Example usage:
mixed_list = [1, "2", 3.14, "hello", True, None, "False", 0, "10"]

type_casting_list_elements(mixed_list)

#Demonstrate with a list containing lists.
mixed_list2 = [1, [2,3], 3.14, "hello", True]
type_casting_list_elements(mixed_list2)
```

```python
#Q.14.  Write a program to reverse a string using a while loop.


def reverse_string_while(input_string):
    """
    Reverses a string using a while loop.

    Args:
        input_string: The string to reverse.

    Returns:
        The reversed string.
    """

    reversed_string = ""
    index = len(input_string) - 1

    while index >= 0:
        reversed_string += input_string[index]
        index -= 1

    return reversed_string

# Example usage:
my_string = "hello world"
reversed_string = reverse_string_while(my_string)
print("Original string:", my_string)
print("Reversed string:", reversed_string)

#Example with an empty string
empty_string = ""
reversed_empty = reverse_string_while(empty_string)
print("Original empty string:", empty_string)
print("Reversed empty string:", reversed_empty)

#Example with a single character string
single_char = "a"
reversed_single = reverse_string_while(single_char)
print("Original single character string:", single_char)
print("Reversed single character string:", reversed_single)
```

```python
#Q.15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.


def factorial_while(number):
    """
    Calculates the factorial of a number using a while loop.

    Args:
        number: The number for which to calculate the factorial.

    Returns:
        The factorial of the number, or an error message if the number is negative.
    """

    if number < 0:
        return "Factorial is not defined for negative numbers."
    elif number == 0:
        return 1
    else:
        factorial = 1
        while number > 0:
            factorial *= number
            number -= 1
        return factorial

# Example usage:
try:
    user_number = int(input("Enter a non-negative integer: "))
    result = factorial_while(user_number)
    print(f"The factorial of {user_number} is: {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")

#Example test cases
print(factorial_while(5))
print(factorial_while(0))
print(factorial_while(-1))
```
